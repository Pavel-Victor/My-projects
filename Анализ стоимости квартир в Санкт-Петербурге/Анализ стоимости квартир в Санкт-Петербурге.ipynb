Три балкона еще можно согласиться, но четыре и пять балконов вызывают большое сомнение. Для начала заменим, согласно логике выданного задания, отсутствующие данные на ноль, и поизучаем кол-во балконов.

```python
data['balcony'] = data['balcony'].fillna(0)
data['balcony'].hist(bins = 10)
```

Из гистограммы видно, что основная масса без балконов, 1 или 2 балкона. Это соответсвует моему жизненному опыту, но посмотрим, на эти данные в строках

```python
display('Пять балконов')
display(data[['rooms','balcony','total_area']].loc[data['balcony'] == 5])
display('Четыре балкона')
display(data[['rooms','balcony','total_area']].loc[data['balcony'] == 4])
display('Три балкона')
display(data[['rooms','balcony','total_area']].loc[data['balcony'] == 3])
```

Многие данные вызывают сомнения. Не может в однокомнатной квартире 52 кв.м быть пять балконов. Максимальное количество балконов , что может быть, это количество комнат плюс еще один. Я думаю, вполне логично, если количество балконов превышает количество комнат, то заменить на количество комнат.



```python
for i in range(1,5):
    data.loc[(data.loc[:,'balcony'] > data.loc[:,'rooms']) & 
             (data.loc[:,'rooms'] == i), 'balcony'] = i
    display('Количество балконов:', i)
    display(data[['rooms','balcony','total_area']].loc[data['balcony'] == i])

```

```python
# создадим датафрайм, куда заведем данные, которые будем использовать для исследования
list = [['площадь квартиры','total_area'],
       ['цена продажи','last_price'],
       ['цена квадратного метра','price_sq'],
       ['количество комнат','rooms'],
       ['высота потолка','ceiling_height']]
df_list = pd.DataFrame(list, columns = ['name','indx'])
display(df_list)



# создаем функцию
# она получает имя столбца, по которому будет строит гистограмму и наименование графика
# вычисляет квартили и диапазон. Для совместимости графиков диапазон используется один, исходя из данных
# по центру, так как там разбросы выше
# затем строятся графики
def build_hist(a,b):
    q1 = spb_centre[a].quantile(0.25)
    q3 = spb_centre[a].quantile(0.75)
    iqr = q3-q1
    q_low = q1-1.5*iqr
    q_hight = q3+1.5*iqr
    spb_centre[a].hist(bins = 30, range = (q_low,q_hight)) 
    plt.title(b+' по центру')
    plt.figure(figsize=(10,15))
    plt.show()
    
    spb_prov[a].hist(bins = 30,range = (q_low,q_hight)) 
    plt.title(b+' по области' )
    plt.figure(figsize=(10,15))
    plt.show()

    
for i in range(len(df_list)):
    a2=df_list.at[i,'name']
    a1=df_list.at[i,'indx']
    build_hist(a1,a2)
    
    
    
        
        
        
```

В целом, высота потолков в центре выше, чем по области. В центре основная масса от 2,7 до 3,5 метров, в области основная масса до 2,75
